# DFT + Hubbard U and Bader charges 

```python
def plot_u(fname="./bands_data/bnds_10.temp",ax=None,label=True,for_second=False):
    p=get_bands(fname)
    c=["#ff595e","#1982c4"]
    ls="--"
    lw=2
    if for_second:c=["","k"];ls="-";lw=1.5
        
    for i in range(len(bp(p).bs_plot_data()['distances'])):
        d=bp(p).bs_plot_data()['distances'][i]
        e_up=bp(p).bs_plot_data()['energy'][i]['1']
        e_down=bp(p).bs_plot_data()['energy'][i]['-1']
        for k in range(len(e_up)):
            if not for_second:
                ax.plot(d,e_up[k],c=c[0],lw=2)
            ax.plot(d,e_down[k],c=c[1],lw=lw,ls=ls)

    for j,i in enumerate(bp(p).bs_plot_data()['ticks']["distance"]):
        ax.axvline(i,c="k",ls="--",lw=.5)
    ax.set_xticks(bp(p).bs_plot_data()['ticks']["distance"])
    ax.set_xticklabels(bp(p).bs_plot_data()['ticks']["label"])
    ax.axhline(0,ls="-.",c="k")
    if not for_second:
        ax.set_ylim(-5,5)
    else:
        ax.set_ylim(-1.6,9.5)
    ax.autoscale(enable=True, axis='x', tight=True)
#     if label:
#         ax.set_ylabel("Energy (eV)")


#---------------dos plot
def plot_dos(ax):
    import sumo.io.questaal as qs
    dos,_=qs.read_dos(tdos_file="./bands_data/dos.temp",gaussian=2e-3)
    c=["#ff595e","#1982c4"]
    up=np.array(dos.as_dict()['densities']['1'])*1.12
    down=np.array(dos.as_dict()['densities']['-1'])*1.12
    energy=dos.as_dict()['energies']
    energy=np.array(energy)-dos.efermi
    ax.plot(-1*up,energy,c=c[0],label="$\\uparrow$")
    ax.fill_betweenx(energy,-1*up,color=c[0],alpha=0.5)
    ax.plot(1*down,energy,c=c[1],ls="--",label="$\\downarrow$")
    ax.fill_betweenx(energy,1*down,color=c[1],alpha=0.5)
    ax.set_ylim(-5,5)
    ax.set_xlim(-40,40)
    ax.axhline(0,c="k",ls="-.")
    ax.set_title("U= 2.04 eV")
#     ax.set_ylabel("Energy (ev)")
    ax.set_xlabel("DOS")
#     ax.legend()
```

```python
fname=["./bands_data/bnds_10.temp","./bands_data/bnds_15.temp","./bands_data/bnds_20.temp","./bands_data/bnds_25.temp"]
U=[0.15,0.20,0.25,0.30]
fig = plt.figure(figsize=(5*(len(fname)+.2),5)) 
gs = gridspec.GridSpec(1,(len(fname)+1), width_ratios=[1/4,1,1,1,1]) 
ax=[]
for i in gs:
    ax.append(plt.subplot(i))
    
plot_dos(ax[0])  
plt.rcParams.update({'font.size': 14})
for i,j in enumerate(fname):
    plot_u(j,ax[i+1])
    ax[i+1].set_title("$U="+str(np.round(13.605662285137*U[i],2))+"$ eV")

ax[0].set_ylabel("Energy (eV)")
    
colors = ["#ff595e","#1982c4"]
ls=["-","--"]
lines = [Line2D([0], [0], color=c, linewidth=3, linestyle=ls[i]) for i,c in enumerate(colors)]
labels = ['$\\uparrow$','$\\downarrow$']
fig.legend(lines, labels, loc = (0.49,.90), ncol=2,fontsize=20 )

plt.tight_layout()
plt.savefig("figs/U-depedence.png",dpi=500, bbox_inches="tight")
plt.show()
```

# Plotting Model Hameltonian plots with projection

```python
from pythtb import *
import matplotlib.pyplot as plt
def plot_fin(evals,vec,k,node=None,cmap="bwr",lw=2,ax=None,transperent=0,alpha=1):
    import numpy as np
    import matplotlib.pyplot as plt
    from matplotlib.collections import LineCollection
    from matplotlib.colors import ListedColormap, BoundaryNorm
    k_node,label=k
    if node is not None:
        index_nums=node
    else:
        index_nums1 = [0]
        index_nums2 = [-1]
        index_nums = index_nums1+index_nums2
    colors=[]
    for j in range(vecs.shape[0]):
        col=[]
        for i in range(len(k_dist)):
            col.append(np.linalg.norm(vecs[j,i,:][index_nums], ord=2))
        colors.append(col)

    
    def make_segments(x, y):
        points = np.array([x, y]).T.reshape(-1, 1, 2)
        segments = np.concatenate([points[:-1], points[1:]], axis=1)

        return segments
    def clear_frame(ax=None): 
        # Taken from a post by Tony S Yu
        if ax is None: 
            ax = plt.gca() 
        ax.xaxis.set_visible(False) 
        ax.yaxis.set_visible(False) 
        for spine in ax.spines.itervalues(): 
            spine.set_visible(False) 
    def colorline(x, y, z=None, cmap=plt.get_cmap(cmap), norm=plt.Normalize(0.0, 1.2), linewidth=2, alpha=1.0):

        if z is None:
            z = np.linspace(0.0, 1.0, len(x))

        if not hasattr(z, "__iter__"):  # to check for numerical input -- this is a hack
            z = np.array([z])

        z = np.asarray(z)

        segments = make_segments(x, y)
        lc = LineCollection(segments, array=z, cmap=cmap, norm=norm, linewidth=linewidth, alpha=alpha)

        ax = plt.gca()
        ax.add_collection(lc)

        return lc

    x = k_dist
    if transperent:
        from matplotlib.colors import LinearSegmentedColormap
        ncolors = 256
        color_array = plt.get_cmap(cmap)(range(ncolors))

        # change alpha values
        color_array[:,-1] = np.linspace(1.0,0.0,ncolors)

        # create a colormap object
        map_object = LinearSegmentedColormap.from_list(name='cmap_alpha',colors=color_array)

        # register this new colormap with matplotlib
        plt.register_cmap(cmap=map_object)
        cmap='cmap_alpha'

    if ax==None:
        fig, ax = plt.subplots(figsize=(9,3),facecolor="w")
    for i in range(vecs.shape[0]):

        y = evals[i]



        colorline(x, y,z=colors[i],alpha=alpha,linewidth=lw,cmap=cmap)

    ax.set_xlim(x.min(), x.max())
    ax.set_ylim(evals.min(), evals.max())
    ax.axhline(np.max(evals[int(evals.shape[0]/2)-1]),c="k",linestyle="--",linewidth=1,alpha=0.5)
    
    ax.set_xticks(k_node, label)
    for n in range(len(k_node)):
          ax.axvline(x=k_node[n], linewidth=0.5, color='k')
    ax.set_xticks(k_node)
    ax.set_xticklabels(label)

def get_model():
    lat=[[1.0,0,0],[1/2,np.sqrt(3)/2,0],[0,0,2]]
    orb=[[0,0,0],[0,0,.2],[0,0,.5],[0,0,.7]]

    num_layers=10
    t1=.5
    t2=2
    t3=t1
    t4=t2
    tli=-.6
    tco=.09
    u=(np.abs(tli)+np.abs(tco))*2.6
    my_model=tb_model(3,3,lat,orb)
    #t1=t2=t3=t4=0
    my_model.set_hop(t1, 0, 1, [0,0,0])
    my_model.set_hop(t2, 1, 2, [0,0,0])
    my_model.set_hop(t3, 2, 3, [0,0,0])
    my_model.set_hop(t4, 3, 0, [0,0,1])


    t11=tli
    my_model.set_hop(t11, 0, 0, [0,1,0])
    my_model.set_hop(t11, 1, 1, [0,1,0])

    my_model.set_hop(t11, 0, 0, [1,0,0])
    my_model.set_hop(t11, 1, 1, [1,0,0])

    my_model.set_hop(t11, 0, 0, [1,-1,0])
    my_model.set_hop(t11, 1, 1, [1,-1,0])

    t11=tco
    my_model.set_hop(t11, 2, 2, [0,1,0])
    my_model.set_hop(t11, 3, 3, [0,1,0])

    my_model.set_hop(t11, 2, 2, [1,0,0])
    my_model.set_hop(t11, 3, 3, [1,0,0])

    my_model.set_hop(t11, 2, 2, [1,-1,0])
    my_model.set_hop(t11, 3, 3, [1,-1,0])

    my_model.set_onsite([u,u,-u,-u])
    return my_model
```

```python
num_layers=1
my_model=get_model()
model=my_model.cut_piece(num_layers,2,glue_edgs=False)
path=[[0.0,0.0],[.5,0],[2/3.,1/3],[0.0,0]]
label=(r'G',r'$M$', r'$K$', r'G')
(k_vec,k_dist,k_node)=model.k_path(path,81,report=0)
evals,vecs=model.solve_all(k_vec,eig_vectors=True)

fig,ax=plt.subplots(1,2,facecolor="w",figsize=(9,4))
plot_fin(evals,vecs,(k_node,label),node=[0,1],cmap="RdBu_r",lw=1.5,ax=ax[1])
colors = ["#ff595e","#1982c4"]
ls=["-","-"]
lines = [Line2D([0], [0], color=c, linewidth=2, linestyle=ls[i]) for i,c in enumerate(colors)]
labels = ['Li','CoO$_2$']
ax[1].legend(lines, labels, ncol=2,fontsize=14 ,loc="center")

plot_u(fname[0],ax[0],for_second=True)

for i in ax:
    i.set_ylabel("Energy (eV)")
    
ax[0].set_title("DFT spin $\\downarrow$ (U=2.04 eV)")
ax[1].set_title("$H_{eff}$")
plt.tight_layout()
plt.savefig("figs/monolayer.png",dpi=500, bbox_inches="tight")
plt.show()
```
