```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.loadtxt("eps_rpa_licoo2.csv",delimiter=",")
e=data.T[0]
nlf=data.T[1]+1j*data.T[2]
lf=data.T[3]+1j*data.T[4]
```

```python
fig,ax=plt.subplots(figsize=(8,4))
plt.plot(e,np.imag(nlf),label="w/o Local Field",color="k",ls="--")
plt.plot(e,np.imag(lf),label="w Local Field",color="k",ls="-")
plt.legend()
plt.show()
```
